```python
import random
def Normal():
    print("Like a normal one huh?? Great Choice!")
    print('''Below are the rules and regulations :
    1. You will be competing against the computer which is completely unbiased.
    2. You will be provided with a box in which you have to enter any number between 1 to 10.
    3. Just like you, computer will also generate any random integer between 1 to 10.
    4. If your's choice differs with the computer's choice, BINGO! score will be updated!
    5. If your's choice is same as the computer's choice, SORRY! You are out!
    6. You also have the choice to bat for n number of balls between the range 1 to 60 i.e. a maximum of 10 overs.
    7. Your final score along with wickets will be displayed after every ball!
    8. After fall of 10 wickets your game will automatically be terminated.''')
    
    balls=int(input('Enter no. of balls you want to play :'))
    if 0>balls or 60<balls:
        print('Kindly enter no. of balls as per rules give nabove')
    else:
        print('Get your sleeves up dude!')
    n_score=0
    n_wickets=0
    c_score=0
    c_wickets=0
    for ball in range(balls):
        n=int(input('Enter any number between 1 to 10 :'))
        c=random.randint(1,10)
        if n==c:
            n_wickets+=1
            c_wickets+=1
            print('''Oops! You are out!''')
            print('Your score :',n_score,'and wickets are :',n_wickets)
        elif n!=c:
            n_score+=n
            c_score+=c
            print('Your score :',n_score,'and wickets are :',n_wickets)
        if n_wickets==10:
            break
    print('Your score :',n_score,'/',n_wickets,'and Computers score :',c_score,'/',c_wickets)
    if n_score>c_score:
        print('Hurray! You won against the computer!')
    elif n_score<c_score:
        print('Sorry! You lost the game. Better luck next time!')
    elif n_score==c_score:
        print('It was a draw!')
        
Normal()
```

```python
def Delux():
    print("You ready for some real shit?? Show us your inner cricketer!")
    print('''Below are the rules and regulations :
    1. You will be competing against the computer which is completely unbiased.
    2. You will be provided with a box in which you have to enter any number between 1 to 10.
    3. Just like you, computer will also generate any random integer between 1 to 10.
    4. If your's choice differs with the computer's choice
       and your choice is not equal to computer's choice+1 or computer's choice-1 , BINGO! score will be updated!
    5. If your choice is equal to computer's choice+1 or computer's choice-1 , SORRY! you are out!
    6. If your's choice is same as the computer's choice, BINGO! Your score gets squared up!
    ^^Here is an example. Suppose your choice is 5 and computer's choice is either 4 or 6, you will be out. 
      If computer's choice is not 4 or 6 or 5 your score will be added normally. But if your choice is equal to 
      computer's choice i.e. 5 on both sides then 5**2 will be added to the final score^^
    7. You also have the choice to bat for n number of balls between the range 1 to 60 i.e. a maximum of 10 overs.
    8. Your final score along with wickets will be displayed after every ball!
    9. After fall of 10 wickets your game will automatically be terminated.''')
    Balls=int(input('Enter no. of balls you want to play :'))
    if 0>Balls or 60<Balls:
        print('Kindly enter no. of balls as per rules given above')
    else:
        print('Get your sleeves up dude! You have chosen real shit!')
    n1_score=0
    n1_wickets=0
    c1_score=0
    c1_wickets=0
    for Ball in range(Balls):
        n1=int(input('Enter any number between 1 to 10 :'))
        c1=random.randint(1,10)
        if n1==c1:
            print('BINGO!')
            n1_score+=n1**2
            c1_score+=c1**2
            print('Your score :',n1_score,'and wickets are :',n1_wickets)
        elif n1==(c1-1) or n1==(c1+1):
            n1_wickets+=1
            c1_wickets+=1
            print('Oops! You are out!')
            print('Your score :',n1_score,'and wickets are :',n1_wickets)
        else:
            n1_score+=n1
            c1_score+=c1
            print('Your score :',n1_score,'and wickets are :',n1_wickets)
        if n1_wickets==10:
            break
    
    print('Your score :',n1_score,'/',n1_wickets,'and Computers score :',c1_score,'/',c1_wickets)
    if n1_score>c1_score:
        print('Hurray! You won against the computer!')
    elif n1_score<c1_score:
        print('Sorry! You lost the game. Better luck next time!')
    elif n1_score==c1_score:
        print('It was a draw!')

Delux()
```
